## Pakia Data ya Titanic na Takwimu


Tutapakia data halisi ya abiria wa Titanic na kuhesabu takwimu.

Kisha tutajenga injini ya majibu ambayo inajibu maswali kwa Kiswahili kwa kutumia **nambari halisi** kutoka kwenye data.

````python
# --- Pakia Data ya Titanic ---

try:
    url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
    titanic_df = pd.read_csv(url)
    print(f'Data ya Titanic imepakiwa: abiria {titanic_df.shape[0]}')
except Exception:
    titanic_df = None
    print('Imeshindwa kupakia data --- mazungumzo yatatumia ujuzi wa AI tu.')

def pata_takwimu():
    """Hesabu takwimu halisi kutoka kwenye data."""
    if titanic_df is None:
        return None
    df = titanic_df
    umri_df = df.dropna(subset=['Age'])
    return {
        'jumla':       df.shape[0],
        'walionusurika': int(df['Survived'].sum()),
        'kiwango':     df['Survived'].mean(),
        'wanawake':    df[df['Sex'] == 'female']['Survived'].mean(),
        'wanaume':     df[df['Sex'] == 'male']['Survived'].mean(),
        'daraja_1':    df[df['Pclass'] == 1]['Survived'].mean(),
        'daraja_2':    df[df['Pclass'] == 2]['Survived'].mean(),
        'daraja_3':    df[df['Pclass'] == 3]['Survived'].mean(),
        'watoto':      umri_df[umri_df['Age'] < 18]['Survived'].mean(),
        'watu_wazima': umri_df[umri_df['Age'] >= 18]['Survived'].mean(),
        'umri_wastani': umri_df['Age'].mean(),
        'nauli_wastani': df['Fare'].mean(),
    }

takwimu = pata_takwimu()
if takwimu:
    print(f'\nTakwimu za Titanic:')
    print(f'   Kiwango cha kunusurika: {takwimu["kiwango"]:.1%}')
    print(f'   Wanawake: {takwimu["wanawake"]:.1%} | Wanaume: {takwimu["wanaume"]:.1%}')
    print(f'   Daraja 1: {takwimu["daraja_1"]:.1%} | Daraja 2: {takwimu["daraja_2"]:.1%} | Daraja 3: {takwimu["daraja_3"]:.1%}')
    print(f'   Watoto: {takwimu["watoto"]:.1%} | Watu wazima: {takwimu["watu_wazima"]:.1%}')
print()
print('Safu ya uchambuzi wa data iko tayari')
````

````python
# --- Injini ya Majibu kutoka Data (kwa Kiswahili) ---

def jibu_kutoka_data(swali: str) -> str:
    """Jaribu kujibu kwa kutumia takwimu halisi za data."""
    if not takwimu:
        return None
    s = swali.lower()
    
    if ('wanawake' in s or 'jinsia' in s) and ('wanaume' in s or 'jinsia' in s):
        return (f'DATA HALISI: Wanawake walinusurika kwa {takwimu["wanawake"]:.1%} '
                f'dhidi ya wanaume kwa {takwimu["wanaume"]:.1%}. '
                f'Sera ya "wanawake na watoto kwanza" ilitekelezwa wazi.')
    
    if ('daraja' in s or 'tabaka' in s) and ('nusurik' in s or 'athir' in s or 'kiwango' in s):
        return (f'DATA HALISI: Daraja 1: {takwimu["daraja_1"]:.1%}, '
                f'Daraja 2: {takwimu["daraja_2"]:.1%}, '
                f'Daraja 3: {takwimu["daraja_3"]:.1%}. '
                f'Daraja la juu = karibu na mashua za kuokoa.')
    
    if 'umri' in s or 'watoto' in s or 'mtoto' in s:
        return (f'DATA HALISI: Watoto (chini ya 18): {takwimu["watoto"]:.1%} kunusurika '
                f'dhidi ya watu wazima: {takwimu["watu_wazima"]:.1%}. '
                f'Umri wa wastani: miaka {takwimu["umri_wastani"]:.0f}.')
    
    if 'kiwango' in s or 'jumla' in s or 'wangapi' in s:
        return (f'DATA HALISI: Kiwango cha kunusurika kilikuwa {takwimu["kiwango"]:.1%} '
                f'({takwimu["walionusurika"]}/{takwimu["jumla"]} abiria).')
    
    if 'nauli' in s or 'tiketi' in s or 'bei' in s or 'gharama' in s:
        return (f'DATA HALISI: Nauli ya wastani ilikuwa pauni {takwimu["nauli_wastani"]:.1f}. '
                f'Nauli kubwa ilihusiana sana na kunusurika.')
    
    return None

print('Injini ya majibu kwa Kiswahili iko tayari')
````

## Mazungumzo na AI kwa Kiswahili


Hapa ndipo Generative AI inang'aa zaidi kuliko Predictive AI.

Badala ya kupata 0 au 1 tu, tunaweza **kuzungumza na data** kwa Kiswahili:
- Uliza *kwa nini* kitu kilitokea
- Pata ufahamu kwa lugha rahisi
- Chunguza hali za "vipi kama"
- Tumia AI kama **mshirika wa uchambuzi**

### Jinsi Inavyofanya Kazi
1. Maswali ya data --- tunajibu kwa **nambari halisi** kwanza
2. Maswali ya ujuzi --- tunaomba AI ijibu **kwa Kiswahili**
3. Kama yote yanashindwa --- tuna **majibu ya akiba**

> *Tabaka tatu za kuaminika: Data > AI > Akiba*

````python
# --- Kazi ya Mazungumzo kwa Kiswahili ---
# Inajaribu: Data > Ollama > Gemini > Akiba

def zungumza(swali: str) -> str:
    """Uliza swali kuhusu Titanic kwa Kiswahili."""
    
    # Tabaka 1: Data halisi
    jibu_data = jibu_kutoka_data(swali)
    if jibu_data:
        return jibu_data
    
    # Tabaka 2: AI (Ollama kwanza kwa faragha, kisha Gemini)
    ai_prompt = (
        f'{KISWAHILI_SYSTEM}\n\n'
        f'Ukweli muhimu: Abiria 891, 38% walinusurika. '
        f'Wanawake: 74%, Wanaume: 19%. '
        f'Daraja 1: 63%, Daraja 2: 47%, Daraja 3: 24%.\n\n'
        f'Swali: {swali}\n\n'
        f'Jibu kwa Kiswahili sanifu, kwa ufupi na uwazi:'
    )
    
    # Jaribu Ollama kwanza (ndani, faragha)
    if ollama_ready:
        try:
            resp = requests.post(
                f'{OLLAMA_URL}/api/generate',
                json={'model': ollama_model, 'prompt': ai_prompt, 'stream': False, 'options': {'temperature': 0.3}},
                timeout=30
            )
            if resp.status_code == 200:
                maandishi = resp.json().get('response', '').strip()
                if maandishi:
                    return f'Ollama: {maandishi}'
        except Exception:
            pass
    
    # Jaribu Gemini
    if gemini_ready:
        try:
            response = gemini_model.generate_content(ai_prompt)
            maandishi = response.text.strip()
            if maandishi:
                return f'Gemini: {maandishi}'
        except Exception:
            pass
    
    # Tabaka 3: Majibu ya akiba
    return ('Kulingana na rekodi za historia: Kunusurika kulitegemea jinsia, daraja, na umri. '
            'Wanawake na watoto wa matabaka ya juu walikuwa na nafasi bora zaidi.')

print('Mfumo wa mazungumzo kwa Kiswahili uko tayari!')
print()
print('Matumizi: zungumza("swali lako hapa")')
print()
print('Jaribu hizi:')
print('  zungumza("Kwa nini wanawake walinusurika zaidi kuliko wanaume?")')
print('  zungumza("Daraja liliathiri vipi kunusurika?")')
print('  zungumza("Kiwango cha kunusurika kilikuwa kipi?")')
print('  zungumza("Tunajifunza nini kutoka kwa msiba wa Titanic?")')
````

## Majaribio ya Prompt Engineering kwa Kiswahili


Sasa tutajaribu jinsi **mabadiliko madogo kwenye prompt** yanavyoleta **matokeo tofauti kabisa**.

Hii ndiyo ujuzi muhimu wa kufanya kazi na Generative AI  na inafanya kazi kwa Kiswahili pia!

````python
# --- Jaribio 1: Jukumu Linabadilisha Kila Kitu ---
# Swali moja, majukumu tofauti --- angalia jibu linavyobadilika

swali = 'Kwa nini watu wengi walikufa kwenye Titanic?'

majukumu = [
    'Wewe ni mhandisi wa usalama wa meli. Jibu kwa Kiswahili.',
    'Wewe ni mwanahistoria anayesoma ukosefu wa usawa wa kitabaka. Jibu kwa Kiswahili.',
    'Wewe ni mtoto wa miaka 10 anayeeleza kwa rafiki yake. Jibu kwa Kiswahili.',
]

print('JARIBIO 1: Jukumu Linabadilisha Jibu')
print(f'Swali: {swali}')
print('=' * 60)

for jukumu in majukumu:
    prompt = f'{jukumu}\n\n{swali}\n\nJibu kwa sentensi 2-3:'
    print(f'\nJukumu: {jukumu[:60]}...')
    
    if ollama_ready:
        try:
            resp = requests.post(
                f'{OLLAMA_URL}/api/generate',
                json={'model': ollama_model, 'prompt': prompt, 'stream': False, 'options': {'temperature': 0.3}},
                timeout=30
            )
            if resp.status_code == 200:
                print(f'   {resp.json().get("response", "").strip()[:300]}')
                continue
        except Exception:
            pass
    
    if gemini_ready:
        try:
            response = gemini_model.generate_content(prompt)
            print(f'   {response.text.strip()[:300]}')
            continue
        except Exception:
            pass
    
    print('   (Hakuna AI inayopatikana --- jaribu wakati Ollama au Gemini imeandaliwa)')

print('\nSwali moja, mitazamo tofauti kabisa --- kwa kubadilisha jukumu tu!')
````

````python
# --- Jaribio 2: Muundo Unadhibiti Matokeo ---
# Maudhui sawa, muundo tofauti wa jibu

msingi = 'Eleza sababu kuu zilizobainisha kunusurika kwenye Titanic. Jibu kwa Kiswahili.'

miundo = [
    ('Vidokezo',  f'{msingi} Tumia vidokezo (bullet points).'),
    ('Shairi',    f'{msingi} Jibu kama shairi fupi la Kiswahili.'),
    ('Ujumbe',    f'{msingi} Jibu kwa ujumbe mmoja mfupi (chini ya herufi 280).'),
    ('Kwa Mtoto', f'{msingi} Eleza kama unaeleza kwa mtoto wa miaka 5.'),
]

print('JARIBIO 2: Muundo Unadhibiti Matokeo')
print('=' * 60)

for lebo, prompt in miundo:
    print(f'\nMuundo: {lebo}')
    
    if ollama_ready:
        try:
            resp = requests.post(
                f'{OLLAMA_URL}/api/generate',
                json={'model': ollama_model, 'prompt': prompt, 'stream': False, 'options': {'temperature': 0.5}},
                timeout=30
            )
            if resp.status_code == 200:
                print(f'   {resp.json().get("response", "").strip()[:300]}')
                continue
        except Exception:
            pass
    
    if gemini_ready:
        try:
            response = gemini_model.generate_content(prompt)
            print(f'   {response.text.strip()[:300]}')
            continue
        except Exception:
            pass
    
    print('   (Hakuna AI inayopatikana --- jaribu wakati Ollama au Gemini imeandaliwa)')

print('\nMaagizo ya muundo yanabadilisha kabisa jinsi AI inavyopanga jibu lake!')
````

````python
# --- Jaribio 3: Joto = Kiwango cha Ubunifu ---
# Joto la chini = makini, la kuaminika
# Joto la juu = ubunifu, tofauti

prompt = 'Eleza usiku wa mwisho wa Titanic kwa aya moja. Jibu kwa Kiswahili.'

viwango_joto = [0.0, 0.5, 1.0]

print('JARIBIO 3: Joto = Kiwango cha Ubunifu')
print(f'Prompt: {prompt}')
print('=' * 60)

for joto in viwango_joto:
    lebo = {0.0: 'Makini/Ukweli', 0.5: 'Uwiano', 1.0: 'Ubunifu/Tofauti'}[joto]
    print(f'\nJoto: {joto} ({lebo})')
    
    if ollama_ready:
        try:
            resp = requests.post(
                f'{OLLAMA_URL}/api/generate',
                json={'model': ollama_model, 'prompt': prompt, 'stream': False, 'options': {'temperature': joto}},
                timeout=30
            )
            if resp.status_code == 200:
                print(f'   {resp.json().get("response", "").strip()[:400]}')
                continue
        except Exception:
            pass
    
    if gemini_ready:
        try:
            response = gemini_model.generate_content(
                prompt,
                generation_config=genai.types.GenerationConfig(temperature=joto)
            )
            print(f'   {response.text.strip()[:400]}')
            continue
        except Exception:
            pass
    
    print('   (Hakuna AI inayopatikana --- jaribu wakati Ollama au Gemini imeandaliwa)')

print('\nJoto 0 = salama na kutabirika. Joto 1 = ubunifu na kushangaza.')
print('Kwa kazi za ukweli, tumia joto la chini. Kwa kazi za ubunifu, panda juu.')
````

## Muhtasari na Kufunga

---

### Ulimwengu Mbili za AI

| | Predictive AI | Generative AI |
|---|---|---|
| **Ingizo** | Data iliyopangwa (safu na nguzo) | Lugha ya kawaida (prompts) |
| **Tokeo** | Kategoria au nambari | Maandishi, msimbo, picha, hoja |
| **Mafunzo** | Unafundisha modeli kwa data yako | Imefundishwa tayari kwa data kubwa |
| **Ujuzi** | Uhandisi wa vipengele, takwimu | Uhandisi wa prompts, tathmini |
| **Nguvu** | Sahihi, inayopimika | Rahisi kubadilika, ya ubunifu |
| **Kikwazo** | Inatabiri tu ilichofundishwa | Inaweza kudanganya, ngumu kuthibitisha |
| **Bora kwa** | Uainishaji, utabiri | Uundaji wa maudhui, uchambuzi, mazungumzo |

**Hakuna iliyo bora kuliko nyingine --- zinatatua matatizo tofauti.** Miradi bora inachanganya zote mbili.

### Ulichojifunza Leo

1. **Predictive AI** inafuata mfuatano: Pakia > Safisha > Fundisha > Tabiri
2. **Generative AI** inafuata mazungumzo: Prompt > Tengeneza > Tathmini > Boresha
3. **Prompt engineering** ni uprogramu mpya --- jukumu, muktadha, kazi, na muundo ni muhimu
4. **AI ya ndani** (Ollama kwenye Ubuntu) inakupa faragha na udhibiti
